# Data loading

```python
import copy
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
jar = pd.read_csv('clean_data.csv')
jar
```

```python
jar['FD'].unique()
```

```python
fdData = [[] for _ in range(7)]
fdVals = jar['FD'].unique()

for i, val in enumerate(fdVals):
    fd = jar[jar['FD'] == val]
    fd = fd.drop(['FD', 'fuel_gas_swelling'], axis=1)
    fdData[i] = fd
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, sharey=True, figsize=(30, 20))

for i, ax in enumerate(axes.flatten()):
    sns.heatmap(fdData[i].corr(), ax=ax, annot=True, cmap='coolwarm')

plt.tight_layout()
plt.show()
```
