```python
import numpy as np
import pandas as pd
```

```python
def clean(paramFile, swellFile, outFile):
    param = pd.read_csv(paramFile)
    jar = pd.read_csv(swellFile)
    
    foo = []
    for index, row in jar.iterrows():
        for i in range(1, 9):
            new_row = {
                'dGrainHBS': param.iloc[index]['dGrainHBS'],
                'FaceCovMax': param.iloc[index]['FaceCovMax'],
                'SwellLink': param.iloc[index]['SwellLink'],
                'vResol': param.iloc[index]['vResol'],
                'DatomFissGBx': param.iloc[index]['DatomFissGBx'],
                'fNucleate': param.iloc[index]['fNucleate'],
                'aAtomDifFiss': param.iloc[index]['aAtomDifFiss'],
                'FD': row[f'FD{i}'],
                'fuel_swelling': row[f'fuel_swelling{i}'],
                'fuel_gas_swelling': row[f'fuel_gas_swelling{i}']
            }
            foo.append(new_row)
    
    ndf = pd.DataFrame(foo)
    ndf_rounded = ndf.map(lambda x: round(x, 5 - int(np.floor(np.log10(abs(x)))) - 1) if x!=0 else 0)
    ndf_rounded.to_csv(outFile, index=False)

    return True
```

```python
param_file = 'DART_parameters.csv'
swell_files = [
    'hi_FR_hi_GS_160C.csv',
    'hi_FR_hi_GS_200C.csv',
    'hi_FR_lo_GS_160C.csv',
    'hi_FR_lo_GS_200C.csv',
    'lo_FR_hi_GS_100C.csv',
    'lo_FR_hi_GS_160C.csv',
    'lo_FR_lo_GS_100C.csv',
    'lo_FR_lo_GS_160C.csv'
]
```

```python
for swf in swell_files:
    clean(param_file, swf, f'clean_{swf}')
```
