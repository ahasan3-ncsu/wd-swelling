# Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
jar = pd.read_csv('clean_hi_FR_hi_GS_160C.csv')
```

```python
outie = jar[jar['fuel_swelling'] >= 200]
normie = jar[jar['fuel_swelling'] < 200]
```

```python
plt.scatter(jar['vResol'], jar['fuel_swelling'])
```

```python
plt.scatter(outie['vResol'], outie['fuel_swelling'])
```

```python
plt.scatter(normie['vResol'], normie['fuel_swelling'])
```

```python
outie.to_csv('outliers.csv', index=False)
normie.to_csv('normal.csv', index=False)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(jar.corr(), annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(normie.corr(), annot=True, cmap='coolwarm')
```
