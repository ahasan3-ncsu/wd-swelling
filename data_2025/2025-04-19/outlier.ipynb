# Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
jar = pd.read_csv('clean_hi_FR_hi_GS_160C.csv')
```

```python
cutoff = 300
outie = jar[jar['fuel_swelling'] >= cutoff]
normie = jar[jar['fuel_swelling'] < cutoff]

print(len(outie), len(normie), len(outie)/len(normie))
```

```python
plt.figure(figsize=(5,4))
plt.scatter(jar['vResol'], jar['fuel_swelling'])
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(outie['vResol'], outie['fuel_swelling'])
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(normie['vResol'], normie['fuel_swelling'])
plt.show()
```

```python
outie.to_csv('outliers.csv', index=False)
normie.to_csv('normal.csv', index=False)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(jar.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(normie.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(outie.corr(), annot=True, cmap='coolwarm')
plt.show()
```
