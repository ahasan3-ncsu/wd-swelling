# Data organization

```python
import numpy as np
import pandas as pd
```

```python
jar = pd.read_csv('high_FR_low_GR.csv')
jar
```

```python
jar.describe()
```

```python
jar.columns.tolist()
```

```python
foo = []

for index, row in jar.iterrows():
    for i in range(1, 8):
        new_row = {
            'dGrainHBS': row['dGrainHBS'],
            'FaceCovMax': row['FaceCovMax'],
            'SwellLink': row['SwellLink'],
            'vResol': row['vResol'],
            'DatomFissGBx': row['DatomFissGBx'],
            'fNucleate': row['fNucleate'],
            'aAtomDifFiss': row['aAtomDifFiss'],
            'FD': row[f'FD{i}'],
            'fuel_swelling': row[f'fuel_swelling{i}'],
            'fuel_gas_swelling': row[f'fuel_gas_swelling{i}']
        }
        foo.append(new_row)

ndf = pd.DataFrame(foo)
```

```python
ndf.describe()
```

```python
ndf_rounded = ndf.map(lambda x: round(x, 5 - int(np.floor(np.log10(abs(x)))) - 1))
ndf_rounded.describe()
```

```python
ndf_rounded.to_csv('clean_high_FR_low_GR.csv', index=False)
```
